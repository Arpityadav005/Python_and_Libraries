```python
#Hybrid inheritance >> multiple type of inheritance
class Vehicle:
    def info(self):
        print("This is vehicle")

class Car(Vehicle):
    def car_info(self,name):
        print("this is car info: ",name)

class Truck(Vehicle):
    def truck_info(self,name):
        print("Truck info: ",name)

class SportsCar(Car, Vehicle):
    def sports_info(self):
        print("This is sports car info")
```

```python
class Laptop:
    def __init__(self,brand,price,graphics):
        print("Inside Laptop Class")
        self.brand = brand
        self.price = price
        self.graphics = graphics
    def buy(self):
        print("I am buying a laptop")
class Tab(Laptop):
    def buy(self):
        print("I am buying a tab")
        super().buy()
```

```python
t=Tab("Lenovo","30K","AMD")
t.buy() 
```

```python
#Can i also access the attribute of the attributes of the parent class using super()?
#remember the previous problem : if we have constructor if child class then you will not be able to access
#attribute of parent class
#above problem can be solved using super..

class Laptop:
    def __init__(self,brand,price,graphics):
        print("Inside Laptop Class")
        self.brand = brand
        self.price = price
        self.graphics = graphics
    def buy(self):
        print("I am buying a laptop")
class Tab(Laptop):
    def __init__(self,brand,price,graphics,screensize,ram):
        print("Inside TAB class constructor")
        super().__init__(brand,price,graphics)
        self.screesize = screensize
        self.ram = ram
        print("Inside TAB class constructor")
```

```python
t=Tab("Lenovo","30K","AMD","16","4")
```

```python
#Same Method Behaving Differently >> just an another example

class Teacher:
    def lec(self):
        print("This is teacher lecture")

class Student:
    def lec(self):
        print("This is student's lecture")
```

```python
def parcer(objects):
    for i in objects:
        print(i.lec())
```

```python
obj1 = Teacher()
obj2 = Student()
objects = [obj1,obj2]
```

```python
parcer(objects)
```
